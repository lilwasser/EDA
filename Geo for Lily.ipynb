```python
coords = pd.read_csv("/Lily_NYC_Data.csv")
coords.head()
```

```python
! pip install geopy
from geopy.distance import geodesic
```

```python
def get_distance(x, coord2):
    return geodesic((x["Latitude"], x["Longitude"]), coord2).miles
```

```python
herald = (40.749338, -73.987985)
penn = (40.7506, -73.9935)
eighty_sixth = (40.7777, -73.9516)

coords["dist_to_herald"] = coords.apply(lambda x: get_distance(x, herald), axis = 1)
coords["dist_to_penn"] = coords.apply(lambda x: get_distance(x, penn), axis = 1)
coords["dist_to_86"] = coords.apply(lambda x: get_distance(x, eighty_sixth), axis = 1)
```
